```python
#Bubble sort
num=int(input("enter a number man"))
arr=[]
for i in range(num):
  arr.append(int(input(f"enter a element{i+1}:")))
print(f"This is an original array {arr}")
for i in range(num-1):
  for j in range(0,num-i-1):
    if arr[j]>arr[j+1]:
      arr[j],arr[j+1]=arr[j+1],arr[j]
print(f"This is an sorted array {arr}")
```

```python
# Insertion sort
num=int(input("enter a number man"))
arr=[]
for i in range(num):
  arr.append(int(input ("enter a element{i+1}")))
print(f"This is an original array {arr}")
for i in range(1,num):
  key=arr[i]
  while j>=0 and key<arr[j]:
    arr[j+1]=arr[j]
    j-=1
    arr[j+1]=key
print(f"This is an Insertion sort",arr)





```

```python
#Linear Search
n=int(input("Enter the number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}:")))
key=int(input("\nEnter key element to be found:"))
found=False
for i in range(n):
  if arr[i]==key:
    print(f"Element found at position {i+1}")
    found=True
if found==False:
  print("Element not found")
```

```python
# Binary Search
n = int(input("Enter the number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))
arr.sort()
print("\nSorted list:", arr)
key = int(input("\nEnter key element to be found: "))
low = 0
high = n - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print(f"Element found at position {mid + 1}")
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
if not found:
    print("Element not found buddy!!")
```
