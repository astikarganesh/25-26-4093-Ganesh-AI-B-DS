```python
#Selection sort
import pandas as pd
data={"ID":[101,102,103,104,105],
      "name":["danyeee","hart","riyath","miyono","yousof"],
      "section":["A","B","C","D","E"],
      "marks":[10,20,30,44,75]
      }
df=pd.DataFrame(data)
print(df)
print(df.sort_values(by="marks"))
print("\nlocation 0 to 4\n\n",df.iloc[0:4,0:8])
```

```python
#filtering data
print("marks less then 50")
print(df[df["marks"]<50])
print("\n\nmarks greater then 50\n")
print("\nagain headding\n",df[df["marks"]>50])
print("nmarks greater then or equal to 30",df[df["marks"]>=30])
print("\n\nmarks less then or equla to 40",df[df["marks"]<=40])
```

```python
#data addition
print(df.index)
print("back fill of data\n",df.backfill)
print("\noriginal data\n",(df))
print("\n\nadded preffix ",df.add_prefix("col_"))
print("\n\n added suffix",df.add_suffix("_col"))
print("",df.agg)
```
