```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='green', label='Regression line')
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression (Scikit-learn)")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
data = pd.read_csv(url)
print(data.head())
X = data[['rm']]
y = data['medv']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel("Average Rooms (rm)")
plt.ylabel("Median Value (medv)")
plt.title("Linear Regression using URL Dataset")
plt.legend()
plt.show()
```

```python
# Spam or not spam
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
emails = [
    "Win money now", "Limited offer just for you", "Hello friend how are you",
    "Important meeting tomorrow", "Claim your free prize now", "Let's catch up soon"
]

labels = [1, 1, 0, 0, 1, 0]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.3)
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Spam Detection Accuracy:", accuracy_score(y_test, y_pred))
sample = ["Free money offer"]
sample_vec = vectorizer.transform(sample)
print("Prediction (1=Spam, 0=Not):", model.predict(sample_vec)[0])
```

```python
#disease or no disease
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Example dataset: [age, blood_pressure, cholesterol]
X = np.array([
    [25, 120, 200],
    [45, 140, 240],
    [35, 130, 230],
    [50, 150, 260],
    [23, 115, 180],
    [40, 135, 220]
])

y = np.array([0, 1, 1, 1, 0, 1])  # 1 = Disease, 0 = No Disease

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Model
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

print("Disease Prediction Accuracy:", accuracy_score(y_test, y_pred))

# Test new patient
new_patient = [[30, 128, 210]]
print("Prediction (1=Disease, 0=No):", model.predict(new_patient)[0])
```

```python
#(c) Sentiment Analysis (Positive vs Negative)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
reviews = [
    "I love this product", "This is amazing", "Very bad experience",
    "I hate this", "Absolutely fantastic", "Not good at all"
]

labels = [1, 1, 0, 0, 1, 0]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(reviews)
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.3)
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Sentiment Accuracy:", accuracy_score(y_test, y_pred))
sample = ["This product is awesome"]
sample_vec = vectorizer.transform(sample)
print("Prediction (1=Positive, 0=Negative):", model.predict(sample_vec)[0])
```
