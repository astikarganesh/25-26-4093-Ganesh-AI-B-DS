```python




# --- Creating Lists ---

my_list = [1, 2, 3, 'hello', True, 3.14]
print(f"Example list: {my_list}")

# --- Accessing Elements ---
print("\n--- Accessing Elements ---")
print(f"First element : {my_list[0]}")
print(f"Last element : {my_list[-1]}")

# --- Slicing Lists ---
print("\n--- Slicing Lists ---")
print(f" {my_list[1:4]}")
print(f" {my_list[:3]}")
print(f": {my_list[3:]}")

# --- Modifying Elements ---

my_list[0] = 100
print(f" {my_list}")

# --- Adding Elements ---

my_list.append('new_item')
print(f" {my_list}")
my_list.insert(1, 'inserted_item')
print(f" {my_list}")

# --- Removing Elements ---


my_list.remove('hello')
print(f"After remove 'hello': {my_list}")
popped_item = my_list.pop()

del my_list[0]
print(f" {my_list}")


```
